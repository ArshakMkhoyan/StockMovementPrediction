```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr 

def stock_price(Company:str, start, end):
    stock=[Company]
    df = pdr.DataReader(stock, 'yahoo', start, end)
    df.columns=['High', 'Low', 'Open', 'Close', 'Volume', 'Adj Close']
    return df 

Company='XOM'
df=stock_price(Company, "1992-12-31", "2018-12-31")
```

```python
from TechnicalIndicators import get_TechnicalIndicators
```

```python
df = get_TechnicalIndicators(df).add_MACD()\
                       .add_StochasticOscillator()\
                       .add_EVM()\
                       .add_ForceIndex()\
                       .add_AverageTrueRange()\
                       .add_MassIndex()\
                       .modify()
```
